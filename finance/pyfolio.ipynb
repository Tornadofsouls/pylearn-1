# 如何使用pyfolio计算指标和画图

```python
import pandas as pd
df = pd.read_csv("./backtraders/orcl-1995-2014.txt")
df.info()
df.set_index('Date', inplace=True)
df.index = pd.to_datetime(df.index)
df.index = df.index.tz_localize('utc')
print('\n info after set index and tz_localized \n')
df.info()
```

```python
returns = df.Close.pct_change()
returns.head(3)
```

```python
import pyfolio as pf
pf.create_returns_tear_sheet(returns, live_start_date='2013-1-1')
```
